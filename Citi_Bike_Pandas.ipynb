```python
# Citi Bike JC Data. Year 2022 from January to May
import pandas as pd
import datetime
import numpy as np
import os 
```

```python
jc_jan_2022 = pd.read_csv("Resources/JC-202201-citibike-tripdata.csv", low_memory=False)
jc_feb_2022 = pd.read_csv("Resources/JC-202202-citibike-tripdata.csv", low_memory=False)
jc_mar_2022 = pd.read_csv("Resources/JC-202203-citibike-tripdata.csv", low_memory=False)
jc_apr_2022 = pd.read_csv("Resources/JC-202204-citibike-tripdata.csv", low_memory=False)
jc_may_2022 = pd.read_csv("Resources/JC-202205-citibike-tripdata.csv", low_memory=False)

# jc_jan_2022
# jc_feb_2022
# jc_mar_2022
# jc_apr_2022
# jc_may_2022
```

```python
jc_jan_2022 = jc_jan_2022.dropna()
jc_feb_2022 = jc_feb_2022.dropna()
jc_mar_2022 = jc_mar_2022.dropna()
jc_apr_2022 = jc_apr_2022.dropna()
jc_may_2022 = jc_may_2022.dropna()
```

```python
# Combined dataset
citibike_jc_jan_to_may_2022 = pd.concat([jc_jan_2022, jc_feb_2022, jc_mar_2022, jc_apr_2022, jc_may_2022])
citibike_jc_jan_to_may_2022
```

```python
# Export file as a CSV
citibike_jc_jan_to_may_2022.to_csv("Resources/JC-2022-JantoMay-citibike-tripdata.csv", index = False)
```
